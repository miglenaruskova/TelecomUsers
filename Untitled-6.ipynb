```python
# 18 Exploratory Data Analysis and Visualizatiz with 3D wich show how Contract related with SeniorCitizen, MonthlyCharges and TotalCharges 
from mpl_toolkits.mplot3d import Axes3D 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
colors = dfclean['Contract'].map({'Two year':0, 'Month-to-month':1, 'One year':2})
ax.scatter(dfclean['SeniorCitizen'],dfclean['MonthlyCharges'],dfclean['TotalCharges'],c=colors);
```
